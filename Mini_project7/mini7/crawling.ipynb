```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import ElementClickInterceptedException
import time

# 웹드라이버 설정 (경로는 본인의 webdriver 경로로 변경)
driver = webdriver.Chrome()

# 웹 페이지 열기
url = 'https://aivle.kt.co.kr/home/ptl/bbs/main?mcd=MC00000058'
driver.get(url)

# 더보기 버튼 클릭
try:
    for i in range(10):
        more_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'a.btn'))
        )
        more_button.click()
        time.sleep(1)
except Exception as e:
    print(f'더보기 버튼을 클릭할 수 없습니다: {e}')
    print('*' * 100)

dates = []
contents = []

# 게시글 리스트 요소 찾기
posts = driver.find_elements(By.CSS_SELECTOR, 'div.td.subject > a')

for index, post in enumerate(posts):
    try:
        # 스크롤하여 요소가 가시적이 되도록 함
        driver.execute_script("arguments[0].scrollIntoView(true);", post)
        time.sleep(1)

        # 게시글이 클릭 가능할 때까지 대기
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable(post))
        
        # 자바스크립트를 사용하여 클릭
        driver.execute_script("arguments[0].click();", post)
        
        time.sleep(2)  # 게시글 로딩 대기 (필요에 따라 조절)
        
        # 게시글 내용 추출
        date = driver.find_element(By.CSS_SELECTOR, 'div.date').text
        content = driver.find_element(By.CSS_SELECTOR, 'div.bo_content').text
        contents.append(content)
        dates.append(date)
        
        # 뒤로가기
        driver.back()
        
        time.sleep(2)  # 뒤로가기 후 페이지 로딩 대기 (필요에 따라 조절)
        
        # # 게시글 리스트 다시 찾기 (뒤로가기 후 새로 로드)
        # posts = driver.find_elements(By.CSS_SELECTOR, 'div.td.subject > a')
    except ElementClickInterceptedException:
        print(f'게시글 {index+1}를 클릭할 수 없습니다: 다른 요소가 가로막고 있습니다.')
    except Exception as e:
        print(f'게시글 {index+1}를 클릭할 수 없습니다: {e}')

# 브라우저 종료
driver.quit()

# DataFrame으로 변환
df = pd.DataFrame({'date' : dates, 'contents': contents})
df.head()
```

```python
df.to_csv('aivleschool_contents.csv', index=False)
```

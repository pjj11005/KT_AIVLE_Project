### 조원들과 상의해서 각자의 인사이트를 도출하고 발표자료를 만들어보세요.

가설 1에 따르면, 택시운송업 종사자 수와 노선 수간에 음의 상관관계가 있으므로 강서구에 노선이 필요하다고 판단했다. 분석 결과를 보고 버스의 수가 적을 수록 택시를 이용하는 사람이 많아진다고 생각해서 이러한 판단을 내렸다.

가설 2에 따르면, 승하차인구수와 노선수 간에는 양의 상관관계가 있으므로 송파구에 노선이 필요하다고 판단했다. 
처음 '구열 총인구이동(유출+유입)에서 20대 ~ 40대 인구가 차지하는 비율이 높을수록 정류장/노선수가 많을 것이다' 이라는 가설을 세우고 분석을 한 결과 유의미한 양의 상관관계가 있어서 20 ~ 40대 비율이 높은 순으로 자치구를 정렬한 후 노선수를 확인했다. 이 때 비율이 높은 편인데도 노선수가 적은 자치구가 있었고 이유를 찾기 위해 자치구별 승하차 총승객수를 확인해보았더니 해당 자치구들은 승하차 총승객수가 적은 것으로 확인되었다. 이를 보고 총승객수가 더 노선수와 관계가 더 클 것으로 생각하고 가설을 다시 세운 후 분석하니, 결과적으로 총승객수와 노선수 간의 유의미한 양의 상관관계가 있어 총승객수가 많지만 노선수가 적은 송파구를 선정했다.
